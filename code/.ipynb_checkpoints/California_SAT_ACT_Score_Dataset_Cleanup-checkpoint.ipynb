```python
# Code:Â 
def string_num (stringy_number):
    string =''
    for i in stringy_number:
        if i == '.':
            string = string + i
        elif i == '%':
            continue
        else:
            string = string + i 
            
    return float(string)/100
```

```python
# Convert to percentage
df_pov_race_scores_final['Poverty'] = df_pov_race_scores_final['Poverty'].apply(string_num)
```

```python
# Remove Duplicate Schools
df_pov_race_scores_final = df_pov_race_scores_final.drop_duplicates(subset = 'schools', keep = False)
```

```python
# Clean Up Column Titles
df_pov_race_scores_final = df_pov_race_scores_final.rename(columns={'Poverty': 'pct_poverty', 'Asian': 'pct_asian','Black': 'pct_black', 'Hispanic': 'pct_hispanic','Multi-Racial': 'pct_mixed', 'Other': 'pct_other_race', 'White': 'pct_white'})
```
