```python
# CREATE A DATA FRAME
df_clean = pd.DataFrame()
df_cleaner = pd.DataFrame()
```

```python
# ASSIGN IMPORTANT COLUMNS TO OUR CLEANER DATA FRAME
df_clean ['SCHOOL'] = race ['SCHOOL']
df_clean ['ETHNIC'] = race ['ETHNIC']
df_clean ['ENR_TOTAL'] = race ['ENR_TOTAL']
df_clean ['GR_12'] = race ['GR_12']
df_clean ['GR_11'] = race ['GR_11']
```

```python
# CHANGE ETHNIC NUMBERS TO WORDS
col         = 'ETHNIC'
conditions  = [ df_clean[col] == 0, df_clean[col] == 1, df_clean[col] == 2, df_clean[col] == 3, df_clean[col] == 4, df_clean[col] == 5, df_clean[col] == 6, df_clean[col] == 7, df_clean[col] == 9]
choices     = [ 'Other', 'Other', 'Asian', 'Other', 'Other', 'Hispanic', 'Black', 'White', 'Multi-Racial' ]
    
df_clean["ETHNICITY"] = np.select(conditions, choices, default=np.nan)
```

```python
# REMOVE ALL SCHOOLS WITH NO GRADE 12 STUDENTS
#df_clean.drop(df_clean[df_clean.GR_12 == 0].index, inplace=True)
```

```python
# REMOVE ALL SCHOOLS WITH NO GRADE 11 STUDENTS
#df_clean.drop(df_clean[df_clean.GR_11 == 0].index, inplace=True)
```

```python
# RESET DATA FRAME INDEX
df_clean.reset_index(drop=True, inplace=True)
```

```python
# CHECK NUMBER OF UNIQUE SCHOOLS
unique_schools = list(set(df_clean['SCHOOL']))
print (f'There are {len(unique_schools)} number of unique schools')
```

```python
# DELETE COLUMNS WE DON'T CARE FOR REALLY
del df_clean['ETHNIC']
del df_clean['GR_12']
del df_clean['GR_11']
```

```python
df_clean.head()
```

```python
# GROUP [SUM] MALE AND FEMALE ETHNICITY
# reference https://jamesrledoux.com/code/group-by-aggregate-pandas

df_cleaner = df_clean.groupby(['SCHOOL', 'ETHNICITY']).agg({'ENR_TOTAL': ['sum', 'mean', 'max']})
df_cleaner.columns = ['enr_sum', 'age_mean', 'age_max']
df_cleaner = df_cleaner.reset_index()
df_cleaner.head()
```

```python
# PIVOT WITH SCHOOLS AS INDEX, COLUMNS AS ETHNICITY AND SUM AS VALUES
# reference https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html

df_cleaner = df_cleaner.pivot(index='SCHOOL', columns='ETHNICITY', values='enr_sum')
df_cleaner = df_cleaner.reset_index()
df_cleaner.head()
```

```python
df_cleaner.SCHOOL[0]
```

```python
# CONVERT VALUE TO PERCENTAGES
# reference https://stackoverflow.com/questions/50820659/compute-row-percentages-in-pandas-dataframe

df_cleaner = df_cleaner.set_index('SCHOOL')
df_cleaner = df_cleaner.div(df_cleaner.sum(axis=1), axis=0)
df_cleaner = df_cleaner.reset_index()
```
